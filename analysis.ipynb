```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt   
```

```python
path = "air_quality.csv"
df = pd.read_csv(path)

df.head(20)
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp'])
station_names = df['station_name'].unique()

for station in station_names:
    
    station_df = df[df['station_name'] == station]
    
    plt.figure(figsize=(16, 8))
    plt.plot(station_df['timestamp'], station_df['aqi'], label=station)
    plt.title(f'Time Series AQI for Station: {station}')
    plt.xlabel('Time')
    plt.ylabel('AQI')
    plt.grid(True)
    plt.tight_layout()
    
    plt.show()
```
